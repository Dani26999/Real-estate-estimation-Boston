### TRATAMIENTO DE MISSING VALUES

```python
def MissingValues(DataSet, MissingValues):
    TamañoX, TamañoY = DataSet.shape
    NanNum = DataSet.isnull().sum().sum()
        
    Atributos = []
    for col in DataSet.columns:
        Atributos.append(col)
   
    if(MissingValues == "DropNan"):  
        es_NaN = DataSet.isnull()
        filas_con_NaN = es_NaN.any(axis=1)
        num_Filas_NaN, y = DataSet[filas_con_NaN].shape
        DataSet = DataSet[np.invert(filas_con_NaN)]
        print(num_Filas_NaN, " filas de datos han sido eliminados")
        return DataSet
    
    elif(MissingValues == "Media"):
        print(NanNum," valores han sido cambiados por su Media")
        return DataSet.fillna(DataSet.mean())
        
    else:
        print(NanNum," valores han sido cambiados por su Moda")
        return DataSet.fillna(DataSet.mode())
    return DataSet
```

```python
dataset_drop = MissingValues(dataset, "DropNan")
dataset_media = MissingValues(dataset, "Media")
dataset_moda = MissingValues(dataset, "Moda")
print("Media: ", dataset.loc[:,"RM"].mean())
print("Moda: ", dataset.loc[:,"RM"].mode())
```

En este punto hemos hecho tres datasets uno donde tratamos los missing values eliminando las filas que los contienen, otro donde substituimos por la media, y por último substituyendo por la moda. He printado la media y la moda del atributo afectado por los missing values, **RM**, y recorriendo el dataset he comprobado que la media es bastante acertada, por lo que lo más seguro que esta sea la política para tratar los valores NaN que se vaya a usar.

### ESTANDARIZACIÓN
A continuación vamos a estandarizar nuestras muestras para asegurar que un atributo que tiene mayores unidades (números mas grandes) que otro, no va a confudir al modelo y asignarle mayor importancia por el simple hecho de tener valores mayores. Con la función ***standarize*** se ha estandarizado el conjunto de atributos indepndientes que forman la X. La limitación que puede tener esta técnica de estandarización es que si la media esta alterada puede incluir errores a la hora de estandarizar, ya que esta técnica se basa en la media para relaizar la estandarización. ### ¿PCA?
A continuación vamos a comprobar si el dataset necesitaría una PCA para obtener mejores predicciones. Una de las ventajas que se obtienen al utilizar una PCA es que se rompen las correlaciones entre variables independeintes (X), a parte de disminuir la complejidad. Como se podía ver en el mapa de correlaciones si se dan casos de alta correlación entre variables independientes por ejemplo: **NOX** con **DIS**, **RM** con **LSTAT**, **DIS** con **AGE**, **DIS** con **INDUS**... Por lo que a primera vista, quizás si que nos irá bien aplicar una PCA con el ojetivo de romper estas correlaciones y evitar que inroduzcan errores en nuestro modelo predictivo, pero realmente tenemos 13 atributos por lo que no es un número muy grande así que alomejor no va aser necesario.

Primeramente vamos a sacar cual sería el número de componentes óptimo que deberíamos aplicar a nuestras muestras.

```python
p = PCA().fit(X)
plt.plot(np.cumsum(p.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

En el gráfico podemos observar el porcentaje de la varianza respecto el numero de componentes. Parece que a partir de los 10 o 11 componentes es cuando se recoge el 100% de la varianza en el modelo. Probaremos aplicando una PCA de 10 y 11 componentes, anque deja de tener un poco de sentido aplicar una PCA donde solo vamos a reducir entre 3 y 2 columnas.

```python
def PCA_transform(x, p):
    pca = PCA(n_components=p)
    return pca.fit_transform(x)
```

```python
X_PCA_10 = PCA_transform(X, 10)
X_PCA_11 = PCA_transform(X,11)
```

En las variables *X_PCA_10* y *X_PCA_11* estan los datasets aplicando una PCA de 10 y 11 componentes respectivamente, he ultilizado la función ***PCA_transform*** para realizar la transformación PCA.

A continuación voy a dividir los 3 datasets (sin PCA, PCA de 10 componentes y PCA de 11 componentes) que vamos a utilizar para crear los primeros modelos y ver como se comportan. El tamaño del dataset de train será del 80% de muestras y el de test del 20% de muestras (test_size = 0.2, de la función ***train_test_split*** que vamos a utilizar).

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(X, Y, test_size=0.2, random_state=0)
x_train_PCA_10, x_test_PCA_10, y_train_PCA_10, y_test_PCA_10 = train_test_split(X_PCA_10, Y, test_size=0.2, random_state=0)
x_train_PCA_11, x_test_PCA_11, y_train_PCA_11, y_test_PCA_11 = train_test_split(X_PCA_11, Y, test_size=0.2, random_state=0)
```

```python
def regression(x, y):
    # Creamos un objeto de regresión lineal de sklearn
    regr = LinearRegression()

    # Entrenamos el modelo
    regr.fit(x, y)
    return regr
```

```python
#Creamos un regresor entrenado con las muestras sin PCA y vemos que mse tiene
lr1 = regression(x_train1, y_train1)
pred1 = lr1.predict(x_test1)
mse1 = mean_squared_error(y_test1, pred1)
print(np.sqrt(mse1))
print(r2_score(y_test1, pred1))
```

```python
#Creamos un regresor entrenado con las muestras con PCA de 10 componentes y vemos que mse tiene
lr_PCA_10 = regression(x_train_PCA_10, y_train_PCA_10)
pred_PCA_10 = lr_PCA_10.predict(x_test_PCA_10)
mse_PCA_10 = mean_squared_error(y_test_PCA_10, pred_PCA_10)
print(np.sqrt(mse_PCA_10))
print(r2_score(y_test1, pred_PCA_10))
```

```python
#Creamos un regresor entrenado con las muestras con PCA de 11 componentes y vemos que mse tiene
lr_PCA_11 = regression(x_train_PCA_11, y_train_PCA_11)
pred_PCA_11 = lr_PCA_11.predict(x_test_PCA_11)
mse_PCA_11 = mean_squared_error(y_test_PCA_11, pred_PCA_11)
print(np.sqrt(mse_PCA_11))
print(r2_score(y_test1, pred_PCA_11))
```

La función ***regression*** crea, entrena y devuelve un modelo de regresión lienal. Hemos utilizado esta funcion para crear tres modelos entreando con las particiones del dataset anteriormente creadas. Con este regresor hemos hecho unas predicciones y valorado su precisión mediante la función ***mean_squared_error*** de sklearn, esta función calcula el error cuadrático medio a partir de las Y de test y lasa Y predecidas, y mediante la función **r2_score**. La metrica r2 podriamos definirla como la razon de semejanza entre las predicciones y los valores reales, si obtubieramos un r2 de 1 querría decir que nuestro modelo predice exactamente el valor esperado. Como vemos el que tiene un error cuadrático menor (y un r2 mayor) es el dataset sin PCA, luego le seguiría el dataset con un PCA de 11 componentes y finalmente el de 11 componentes. Esto nos indica que tendriamos que aumentar el número de componentes para lograr un menor error e intentar igualar o mejorar la predicción del dataset original, no tiene sentido que aumentemos más el PCA ya qu etenemos 13 atributos y auymentar a una PCA de 12 componentes no tiene mucho sentido, además ya podemos ver que máximo vamos a igualar el comportamiento del dataset original, por ello no vamos a realizar una PCA.

### CROSS-VALIDATION
En este punto vamos a realizar corss validaciones con diferentes valores para el k-fold con el objetivo de encontrar el mejor modelo que nos haga las mejores predicciones. Realizaremos cross validación en vez de la partición del dataset en train y test como hemos hecho anteriromente para obtener un mejor resultado y evitar que haya una mala validación y poder detectar correctamente los casos de underfitting y overfitting.

Para ello vamos a probar por diferentes k-fold, que irán de 2 a 10.

```python
k_folds = [2,3, 4, 5, 6, 7, 8, 9, 10]
metrics = ["neg_mean_squared_error", "r2"]
scores = dict()
for k in k_folds:
    scores[k] = cross_validate(LinearRegression(), X, Y, cv = k, scoring = metrics)
print(scores)
```

Como podemos ver el MSE en todas las cross validaciones por los diferentes k-folds probados, siempre hay una iteración en que el MSE aumenta muchsimo respecto las otras iteraciones, en consecuencia también lo hace el R2 de la iteración correspondiente. Esto se debe a que según las muestras que deje para el test la predicción será relamente mala, ya que para esas muestras en concreto no ha aprendido lo suficiente como para poder hacer una correcta predicción. Claramente tenemos un número muy pequeño de muestras para permitir que el modelo generalice suficiente, y segun las muestras con las que entrenemos y luego validemos tendremos una resultado u otro.

```python
var = []
mse = []
for percent in scores:
    for numbers in scores[percent]:
        i = 0
        aux = 0
        if numbers == "test_neg_mean_squared_error":
            var.append(scores[percent][numbers].var())
        for num in np.nditer(scores[percent][numbers]):
            i = i + 1
            if numbers == "test_neg_mean_squared_error":
                aux = aux + abs(num)
            else:
                aux = aux + num
        scores[percent][numbers] = aux/i
        if numbers == "test_neg_mean_squared_error":
            mse.append(aux/i)
```

```python
print(scores)
print(var)
```

Con este código hemos calculado las medias por cada k-fold de las diferentes métricas que obtenemos al realizar la cross validación. Además he recogido en la lista **var** la varianza respecto el mse en cada iteración de una cross-validación por una k-fold determinada. En la lista **mse** guardo la media del error cuadrático por cada k-fold.

```python
#Printamos la varianza respecto cada k-fold
plt.plot(k_folds, var)
plt.ylabel('VARIANZA')
plt.xlabel('K_FOLD')
plt.show()
```

```python
#Printamos el MSE respecto cada k-fold
plt.plot(k_folds, mse)
plt.ylabel('AVERAGE MSE')
plt.xlabel('K_FOLD')
plt.show()
```

En estas dos últimas gáficas podemos ver como la varianza por cada k-fold  es muy grande (primera gráfica). Esto lo podiamos ver a simple vista en el array scores, como hemos comentado anteriromente, donde el MSE (y R2) en una iteración de un k-fold determinado podia variar muchsimo respecto otra iteración del mismo k-fold. Como vemos en la gráfica del MSE (segunda gráfica) a partir de un k-fold de 7 la mejora del modelo es poco significativa.

### SVR
Viendo los malos resultados que se obtienen al itentar hacer una predicción de esta BD con un regresor lineal voy a probar con una regresión con vectores de soporte (SVR). Priemramente realizaremos un *GridSearch*, para encontrar cuales son los mejores parámetros para crear este modelo. Utilizaré el *GridSearch* ya que al tener pocos datos no tiene un coste demasiado elevado.


```python

param_grid = {'C': [1, 5, 10, 50, 100, 110, 200, 1000], 'kernel': ['linear', 'rbf', 'poly', 'sigmoid']}
gdSVR = GridSearchCV(SVR(), param_grid, cv=7)
gdSVR.fit(x_train1, y_train1)
print('Mejores parámetros: ', gdSVR.best_params_)
```

Al ejecutar el *GridSearch* con un cv = 7, que corresponde al k-fold de una cross validación vemos que los mejores parámetros son: una **c** de 100 y el kernel **rbf**. He utilizado un cv de 7 por las conclusiones que habiamos extraido anterirmente de que a partir de una k-fold de 7 la mejora es insignificativa. A continuación vamos a realizar otro *GridSearch* para encontrar los mejores valores para los parámetros **gamma** y **epsilon** para nuestro modelo SVR con kernel **rbf**, lo realizo separado ya que sino el tiempo de ejcución crece demasiado y no necesitamso que realize todas las comprobaciones.

```python
param_grid2 = {'gamma' : [0.1, 0.2, 0.4, 1,2,3,10], 'epsilon' : [0.1, 0.2, 0.3, 0.5,1]}
gdSVR = GridSearchCV(SVR(kernel="rbf", C= 100), param_grid2, cv=7)
gdSVR.fit(x_train1, y_train1)
print('Mejores parámetros: ', gdSVR.best_params_)
```

Con este último *GridSearch* obtenemos que la mejor **gamma** es 0.1 y la mejor **epsilon** 0.1. A continuación vamos a crear un SVR que entrenaremos y testearemos con la partición que habíamos usado anteirormente, 80% para train y 20% para test, y con los mejores parámetros obtenidos.

```python
svr = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svr.fit(x_train1, y_train1)
y_predicted_svr = svr.predict(x_test1)
print(r2_score(y_test1, y_predicted_svr))
```

Vemos que al crear un SVR con los parámetros encontrados por ambos *GridSearch* obtenemos un r2 muy alto, 0.84 aproximadamente. La metrica r2 podriamos definirla como la razon de semejanza entre las predicciones y los valores reales, si obtubieramos un r2 de 1 querría decir que nuestro modelo predice exactamente el valor esperado, en este caso no obtenemos 1 ya que sería practicamente imposible pero estamos muy cerca, por lo que podemos concluir que este modelo hace una predicción bastante acertada. Con una SVR nos alejamos mucho de esos malos resultados que obteniamos mediante un regresor lineal, que como mucho podiamos obtener un r2 de 0'56, un valor muy bajo.

### RANDOM FOREST REGRESSOR
Hemos visto que con un modelo SVR obtenemos muy buenos resultados, ahora vamos a probar el modelo RandomForestRegressor para ver si obtenemos mejores predicciones.
Este modelo esta basado en una serie de árboles de decision que ayudan a hacer la predicción. Primero haremos un *GridSearch* para buscar cual sería el mejor número de árboles para realizar nuestra prediccion.
Vamos a seguir con la partición que habíamos usado anteirormente, 80% para train y 20% para test.

```python
param_grid3 = {'n_estimators' : [50, 60, 80, 100, 110, 130, 150, 200]}
gdRF = GridSearchCV(RandomForestRegressor(), param_grid3, cv=7)
gdRF.fit(x_train1, y_train1)
print('Mejores parámetros: ', gdRF.best_params_)
```

Con el *GridSearch* vemos que el número de árboles óptimo varia dependiendo de la ejecución por lo que no hay mucha diferencia entre un valor o otro por lo que lo fijaremos en 100 que es lo estándard. A continuación vamos a crear 2 RandomForestRegressor, uno dodne el parámetro **bootstrap** será *false* y otro que será *true*. Este parámetro indica si queremos usar todas las muestras para construir cada árbol (bootstrap = false) o si queremos usar subconjuntos aleatorios para entrenar cada árbol (bootstrap = true), de manera que cada árbol se haya especializado en una parte.

```python
no_bootstrap_RF = RandomForestRegressor(n_estimators=100, bootstrap=False)
no_bootstrap_RF.fit(x_train1, y_train1)
print("R2 sin bootstrapping:", no_bootstrap_RF.score(x_test1, y_test1))
```

```python
bootstrap_RF = RandomForestRegressor(n_estimators=100, bootstrap=True)
bootstrap_RF.fit(x_train1, y_train1)
print("R2 con bootstrapping:", bootstrap_RF.score(x_test1, y_test1))
```

Como vemos si no usando la técnica bootstraping obtenemos un r2 de 0'7 aproximadamente y usando bootstrapping obtenemos 0'83 aproxiamadamente. Hay una clara mejoría usando esta técnica, y era de esperar, ya que ahora tenemso diferentes árboles de decision donde cada uno se habrá focalizado en una serie de muestras por lo que hará predicciones más precisas para las muestras que se ajusten a lo que ha aprendido. Juntando todos los arboles de decision como hace el modelo RandomForestRegressor, al final podemos tener en cuenta lo que cada árbol ha predecido y obetner una predicción más precisa.

Hay que remarcar que obtenemos mejores resultados con el modelo basado en SVR, por lo general, ya que la mejora es insignificante y además el RandomForestRegressor dependiendo de la ejecución puede incluso obtener mejores resultados. Podemos concluir que el comportamiento de estos modelos para nuestro dataset son bastante similares y equivalentes, el modelo SVR es más estable ya que parece que no varia su precisión en función de la ejecución, como pasa con el basado en árboles, por contra este último puede llegar a obtener mejores resultados (hatsa un r2 de 0'85).

### ENSAMBLE
En este punto viendo el comportamiento que tienen todos los modelos vamos a intentar juntarlos para ver si obtenemos mejores resultados que si únicamente usamos un modelo basado en SVR o basado en RandomForesRegressor. Para realizar este ensmable vamos a usar el VotingRegressor, que entrena cada modelo con todo el dataset que se le pasa, en este caso lo veo apropiado por el número de muestras que disponemos, además que el RandomForestRegressor ya usa la técnica de entrenar cada uno de sus árboles con particiones del dataset.

```python
lr_e = LinearRegression()
svr_e = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
rfr_e = RandomForestRegressor(n_estimators=100, bootstrap=True)
er_1 = VotingRegressor([('lr', lr_e),('svr', svr_e), ('rf', rfr_e)])
er_1.fit(x_train1, y_train1)
print(er_1.score(x_test1, y_test1))
```

En este VotingRegressor hemos teniedo en cuenta los tres modelos considerados con sus parámetros óptimos, regresor lineal, SVR, y RandomForestRegressor. Como vemos el r2 cae al 0'8 respecto los modelos SVR y RandomForestRegressor, por lo que estamos obteniendo peores resultados.

```python
er_2 = VotingRegressor([('svr', svr_e), ('rf', rfr_e)])
er_2.fit(x_train1, y_train1)
print(er_2.score(x_test1, y_test1))
```

En este otro VotingRegressor solamente hemos tenido en cuenta los modelos SVR y RandomForestRegressor y hemos obtenido mejores resultados que usando ambos de forma independiente, superamos los 0'86. Si comparamos ambos VotingRegressor podemos ver que al icluir el regresor lineal no añadimos mejora, al contrario, estamos añadiendo confusión al modelo, sus malas predicciones no aportan nada más que error en la predicción final. Podemos reafirmar, que un regresor lineal en este dataset con tan pocas muestras no es un buen modelo para realizar predicciones.

No vamos a probar más modelos ya que dificlmente vamos a obtener mejores predicciones dado un dataset de tamaño tan reducido como el que tenemos, ya que poco conocimiento más puede llegar a extraer otro modelo.

### OBTENER EL MEJOR MODELO CROSSVALIDANDO
En este punto hemos visto que usando un VotingRegressor formado por un modelo SVR y un RandomForestRegressor nos da la mejor predicción. A continuación vamos a utilizar el método de cross validation para intentar encontrar el mejor modelo posible. Para hacer la cross validation vamos a usar nuestro dataset de train (el 80% del dataset) y luego haremos una validación final del mejor modelo encontrado usando el dataset de test (el 20%).
Vamos a probar por las k-folds que habíamos probado anteriormente: 2, 3, 4, 5, 6, 7, 8, 9 y 10.

```python
k_folds = [2,3, 4, 5, 6, 7, 8, 9, 10]
scores2 = dict()
for k in k_folds:
    scores2[k] = cross_validate(VotingRegressor([('svr', svr_e), ('rf', rfr_e)]), x_train1, y_train1, cv = k, scoring = ["r2"], return_estimator=True)
print(scores2)
```

En este punto tenemos las crossvalidaciones realizadas y nos falta encontrar en que k-fold y en que iteración se encuntra el mejor estimador. El siguiente código esta dedicado a este objetivo: 

```python
model = {}
model['k'] = -1
model['r2'] = float('-inf')

for k, d in scores2.items():
    for i in d['test_r2']:
        if model['r2'] < i:
            model['r2'] = i
            model['k'] = k
val = scores2[model['k']]['test_r2']
model['i'] = val.argmax()
print(model)
```

Hemos obtenido que el mejor modelo lo encontramos en la iteración 5 (con índice 4) para una k-fold de 10. Este modelo obtiene un r2 de 0'94, muy cerca del 1, cosa que nos indica que ha hecho una predicción casi perfecta. A continuación vamos a realizar la validación final del mejor modelo, con nuestro dataset de test.

```python
best_model = scores2[10]['estimator'][4]
best_model_prediction = best_model.predict(x_test1)
print(r2_score(y_test1, best_model_prediction))
```

Realizando la validación final obtenemos un r2 de 0'86 aproximadamente, por lo que no hemos mejorado practicamente al realizar la cross validación.
